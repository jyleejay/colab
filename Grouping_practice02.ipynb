### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# make gender values to numeric
users['gender'].replace(['F', 'M'],
                        [0, 1], inplace=True)
```

```python
# this could be another option
replace_dict= {'F':0, 'M':1}
print(replace_dict)
aaa['gender']=aaa['gender'].map(replace_dict)
aaa
```

```python
# make a new df added percentile, could not working with .astype(str) + "%" after round(1)
new_users = users.groupby("occupation").gender.value_counts(normalize = True).mul(100).round(1).reset_index(name = "percent")

# left male's percentage only
new_users = new_users.loc[new_users['gender'] == 1]
new_users = new_users.drop("gender", axis = 1)

# sorting
#new_users.percent.sort_values(ascending = False)
new_users = new_users.sort_values(by = ["percent"], ascending = False)

# for add %
new_users['percent'] = new_users['percent'].astype(str) + '%'
new_users
```

```python
# create a function
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

# apply the function to the gender column and create a new column
users['gender_n'] = users['gender'].apply(gender_to_numeric)


a = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100

# sort to the most male
a.sort_values(ascending = False)
```

```python
# make gender values to numeric -> already've done in step5
#users['gender'].replace(['F', 'M'],
#                        [0, 1], inplace=True)

# make a new df added percentile, could not working with .astype(str) + "%" after round(1)
pc_users = users.groupby("occupation").gender.value_counts(normalize = True).mul(100).round(1).astype(str) + '%'

pc_users
```

```python
# create a data frame and apply count to gender
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

# create a DataFrame and apply count for each occupation
occup_count = users.groupby(['occupation']).agg('count')

# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100

# present all rows from the 'gender column'
occup_gender.loc[: , 'gender']
```
